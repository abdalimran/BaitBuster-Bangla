```python
from rapidfuzz import process, fuzz

def fuzzy_deletion(word, string):
    tokens = string.split()
    matches = process.extract(query=word, choices=tokens, scorer=fuzz.WRatio, score_cutoff=60.0, limit=len(tokens))
    idx_matches = [m[2] for m in matches]
    return ' '.join(filter(lambda token_idx: tokens.index(token_idx) not in idx_matches, tokens))
```

```python
fuzzy_deletion(dataset.loc[17994, 'channel_name'], dataset.loc[17994, 'title'])
```

```python
from pandarallel import pandarallel

pandarallel.initialize(progress_bar=True)

dataset['title_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion(row['channel_name'], row['title']), axis=1)
dataset['title_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion(channel_name_translated.get(row['channel_name']), row['title_debiased']), axis=1)
```

```python
dataset['description_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion(row['channel_name'], row['description']), axis=1)
dataset['description_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion(channel_name_translated.get(row['channel_name']), row['description_debiased']), axis=1)
```

```python
 dataset['title_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion("- (চিকিৎসক, ইংল্যান্ড)", row['title_debiased']), axis=1)
 dataset['description_debiased'] = dataset.parallel_apply(lambda row: fuzzy_deletion("- (চিকিৎসক, ইংল্যান্ড)", row['description_debiased']), axis=1)
```

```python
dataset_cleaned = dataset[dataset['title_debiased'].apply(lambda x:len(x.split())>=5) & dataset['description_debiased'].apply(lambda x:len(x.split())>=10)]
```
